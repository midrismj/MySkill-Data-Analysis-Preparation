```python
import pandas as pd
import numpy as np
```

```python
# Load your dataset
data = pd.DataFrame(
          {
              "Negara": ["Indosenia", "Malaysia", "Singapura", "Malaysia", "Thailand", None],
              "A": [1, 2, None, 2, 3, None],
              "B": [5,None, 7, None, None, None],
              "C": [99, 9, None, 9, 11, None],
              "D": [12, 8, 14, 8, 15, None],
          }
    )

# Print the dataset
print("Original dataset:")
print(data)
```

```python
# Dataset description
data.describe()
```

## Missing Value

```python
# Check for missing values in the dataset
print("Missing values in the entire dataset:")
print(data.isnull().sum())

# Identify columns with missing values
print("\nColumns with missing values:")
print(data.columns[data.isnull().any()].tolist())
```

```python
# Delete rows of data with all missing values
drop_row = data.dropna(how='all')
drop_row
```

```python
# Delete rows of data that have missing values
drop_row = data.dropna(how='any')
drop_row
```

```python
# Delete the column where there are 4 missing data values
drop_col = data.dropna(thresh=4, axis='columns')
drop_col
```

```python
# Fill in the missing data values with the specified values
fill_na = data.fillna({'A': 0, 'B': 10, 'C': data['C'].mean()})
fill_na
```

```python
# Print the original column A
print("Original column A:")
print(data["A"])

# Impute missing values with the column A's mean
data_imputed = data["A"].fillna(data["A"].mean())

# Print the imputed column A
print("\nImputed column A:")
print(data_imputed)
```

```python
# Print the original column B
print("Original column B:")
print(data["B"])

# Forward-fill missing values in column B
data_imputed = data["B"].fillna(method="ffill")

# Print the Forward-filled column B
print("\nForward-filled column B:")
print(data_imputed)

# Backward-fill missing values in column B
data_imputed = data["B"].fillna(method="bfill")

# Print the Backward-filled column B
print("\nBackward-filled column B:")
print(data_imputed)
```

```python
# Print the dataset before interpolation
print("Before interpolation:")
print(data)

# Print the dataset after interpolation
print("\nAfter interpolation:")
print(data.interpolate())
```

## Outlier Data

```python
boxplot = data.boxplot(column=['A', 'B', 'C', 'D'])
```

```python
# Delete rows of data with all missing values
drop_row = data.dropna(how='all')
drop_row
```

```python
# Fill in the missin values with 0
fill_na = drop_row.fillna({'A':0, 'B': 0, 'C': 0, 'D':0})
fill_na
```

```python
# IQR Values from Column C
Q1 = fill_na['C'].quantile(0.25)
Q3 = fill_na['C'].quantile(0.75)
IQR = Q3 - Q1
```

```python
print("Nilai Q1:", Q1)
print("\nNilai Q3:", Q3)
print("\nNilai IQR:", IQR)
```

```python
# Determine the upper limit and lower limit values
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR

print("Nilai upper limit:", upper_limit)
print("\nNilai lower limit:", lower_limit)
```

```python
# Data values that exceed the upper limit
fill_na[fill_na['C'] > upper_limit]
```

```python
# 1: Triming outliers
new_df = fill_na[fill_na['C'] < upper_limit]
new_df
```

```python
boxplot = new_df.boxplot(column=['C'])
```

```python
# If the data value is greater than the upper limit, then change it to the upper limit value
# If the data value is smaller than the lower limit, then change it to the lower limit value

new_df_cap = fill_na.copy()
new_df_cap['C'] = np.where(
    new_df_cap['C'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['C'] < lower_limit,
        lower_limit,
        new_df_cap['C']
    )
)
new_df_cap
```

```python
boxplot = new_df_cap.boxplot(column=['A', 'B', 'C', 'D'])
```
